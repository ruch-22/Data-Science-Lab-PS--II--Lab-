```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
url = "https://raw.githubusercontent.com/selva86/datasets/master/Admission.csv"
data = pd.read_csv(url)

print("Dataset Name: Graduate Admission Dataset")
data.head()
```

```python
# Convert chance of admit into binary (0 or 1)
data['Admission'] = (data['Chance of Admit '] > 0.7).astype(int)

# Independent variables
X = data[['GRE Score', 'CGPA', 'University Rating']]

# Dependent variable
y = data['Admission']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=0)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("Accuracy:")
print(accuracy_score(y_test, y_pred))
```

```python
print("\nAccuracy Score:")
print(accuracy_score(y_test, y_pred))

print("\nClassification Report:")
print(classification_report(y_test, y_pred))
```

```python
coefficients = pd.DataFrame(model.coef_, columns=X.columns)
print("Relationship between variables and admission:")
print(coefficients)
```
